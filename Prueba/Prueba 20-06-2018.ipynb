```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (100, 10)
```

```python
olymp = pd.read_csv('dictionary.csv',sep=",",index_col="Country",encoding="utf-8")
olymp
```

```python
#pregunta 1
gdp=olymp[['GDP per Capita']]
gdp.plot(kind='bar')
plt.show()
```

```python
osummer = pd.read_csv('summer.csv',sep=",",index_col="Year",encoding="utf-8")
owinter = pd.read_csv('winter.csv',sep=",",index_col="Year",encoding="utf-8")
```

```python
#pregunta 2
c=osummer['Country']
medal_counts = c.value_counts()
medal_counts.plot(kind='bar')
plt.show()
```

```python
#pregunta 3
type_medal = osummer.pivot_table(index='Country',values='Athlete', columns='Medal', aggfunc='count')
type_medal.plot(kind='bar')
plt.show()
```
